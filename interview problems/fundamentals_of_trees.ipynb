## **What is a Tree?**

A tree is a connected, acyclic graph with a single root, where each node (except the root) has exactly one parent, forming a 
hierarchical structure.

### **Key Properties (You Must Know These)**
	•	Exactly one root
	•	No cycles
	•	Unique path between any two nodes
	•	Every node has:
        •	0 or more children
        •	Exactly 1 parent (except root)

### **Mental Model (Very Important)**

A tree represents hierarchical ownership.

### **Tree vs Real Systems (Senior Insight)**

“In real systems, trees can become graphs if we introduce shared references, like symlinks in file systems.”
