### **Find the most repeated character in string.**

```python
class Solution:
    def most_repeated_character(self, s: str) -> (str, int):
        char_freq = {}
        for char in s:
            if char not in char_freq:
                char_freq[char] = 0
            char_freq[char] += 1
        return max(char_freq, key=char_freq.get), char_freq[max(char_freq, key=char_freq.get)]


sol: Solution = Solution()
sol.most_repeated_character("hello world")
```

### Kadaneâ€™s Algorithm â€“ Maximum Subarray Sum

```python
from typing import List

"""
âœ… Problem Definition

Given an integer array nums, find the maximum possible sum of a contiguous subarray.

Example:
[-2,1,-3,4,-1,2,1,-5,4] â†’ 6 (subarray: [4,-1,2,1])

âŒ Incorrect Solutions
- [1] â†’ 1 (incorrect: should be [-2,1,-3,4,-1,2,1,-5,4])
- [1,2] â†’ 3 (incorrect: should be [-2,1,-3,4,-1,2,1,-5,4])

"""

class Solution:
    def max_subarray_sum(self, nums: List[int]) -> List[int]:
        max_sum:int = float('-inf')
        current_sum:int = 0

        for i in range(len(nums)):
            current_sum = max(nums[i], current_sum + nums[i])
            max_sum = max(max_sum, current_sum)
        
        return max_sum

sol: Solution = Solution()
sol.max_subarray_sum([-2,1,-3,4,-1,2,1,-5,4])
```
